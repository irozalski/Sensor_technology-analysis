### The principle of operation of light sources::

#### LEDs utilize the phenomenon of electroluminescence to emit excess energy in the form of light caused by the jumping of electrons from the N layer to the P layer in a semiconductor chip. 

#### Halogen lamp work on a principle similar to traditional incandescent bulbs. Current flows from the socket and passes through the base of the bulb to the tungsten filament, heating it up. The filament is enclosed in a quartz bulb filled with a noble gas mixed with a small amount of halogen.

#### In a fluorescent lamp, light is emitted by a phosphor excited by ultraviolet radiation generated by a discharge in a tube filled with gas.

#### In an incandescent bulb, the current flowing through a thin spiral (usually made of tungsten) causes it to heat up to a temperature significantly exceeding 2000 K. The tungsten filament glows, producing a stream of light.

### Electromagnetic wave lengths visible to the human eye are in the range of 380-740 nm. The color of the light we see also depends on the wavelength. Other spectral parameters, such as continuity, affect the application of light sources.

# 2. Spectroscopy in quantitative chemical analysis.
### Through an empty flask, 5 flasks filled with methyl orange of different known concentrations and one of unknown concentration were passed with a beam of light. The absorbance of each was calculated, which allowed for the creation of a graph showing the dependence of absorbance on the concentration of the substance and the determination of the molar absorption coefficient using the Lambert-Beer law. Finally, the concentration of the previously unknown flask was calculated.

```python
dict = {'Nr' : ['The intensity of incident light',1, 2, 3, 4, 5, 6, 'X'],
        
        'Concentration' : ['-', 0.000156,  0.0003125, 0.000625, 0.00125, 0.0025, 0.005, 'X'],
        'Intensity' : [244.5, 239.8, 232, 201.2, 117.2, 29.8, 13.8, 76.7],
        }
vials = pd.DataFrame(dict)

display(vials)
```

```python
vials = vials.set_index('Nr')
```

```python
display(vials)
```

***
$\mathbf{\text{Absorbance}}$<br>
***
$$
  A = log\frac{I_{0}}{I}
$$ 
$I_{0}$ = intensity of incident light (in this case, intensity of the light beam that passed through the empty flask)<br>
$I$ = The intensity of light after passing through an absorbing medium.

```python
vials['Absorbance'] = np.log10(vials.loc['The intensity of incident light']['Intensity']/vials['Intensity'])
```

```python
display(vials)
```

```python
x = vials.loc[1:6]['Concentration']
y = vials.loc[1:6]['Absorbance']


plt.xlabel('Concentration [$mol/dm^3$]', fontsize = 15)
plt.ylabel('Absorbance', fontsize = 15)
plt.scatter(x, y, alpha=0.5)


plt.show()
```

```python
x = vials.loc[1:6]['Concentration']
y = vials.loc[1:6]['Absorbance']

x1 = np.asarray(vials.loc[1:4]['Concentration'], dtype='float')

y1 = np.asarray(vials.loc[1:4]['Absorbance'], dtype='float')

m, b = np.polyfit(x1, y1, 1)

plt.xlabel('Concentration [$mol/dm^3$]', fontsize = 15)
plt.ylabel('Absorbance', fontsize = 15)
plt.scatter(x, y, alpha=0.5)

plt.plot(x1, m*x1+b, color='red')

plt.title(f'slope = {round(m, 3)}, intercept = {round(b, 3)}')
plt.show()
```

### The last two measurement points were omitted in the linear regression range because the A = f(molar concentration) graph has a linear dependence only for small concentrations.

***
$\mathbf{\text{Calculation of the concentration of flask X:}}$<br>

***
Lambert-Beer Law:
$$
  A = εlc
$$ 
c – molar concentration of a substance in a solution [mol/dm³]<br>
l – thickness of the absorbing layer (the path
overcomes radiation by passing through a solution).<br>

$$y = ax+b$$<br>
$$a=εl, l=1cm$$<br>
$$a=ε$$<br>
$$a=292,75 [dm3 * mol^{-1} * cm^{-1}]$$

 

***
$\mathbf{\text{Concentration of flask X:}}$<br>

***

$$
  A_{x} = 0,503
$$ 

$$c=\frac{A_{x}}{al}$$

$$c=\frac{0,503}{292,75}$$

$$c = 1,7*10^{-3}\frac{mol}{dm^3}$$<br>
