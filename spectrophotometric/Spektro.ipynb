## Wstęp teoretyczny

### Podczas ćwiczenia użyte zostały następujące żródła światła:

#### -LED o barwie ciepłej 
#### -LED o barwie zimnej 
#### -Lampa halogenowa
#### -Świetlówka o barwie ciepłej
#### -Świetlówka o barwie zimnej 
#### -Żarówka

### Zasada działania oraz emitowane widmo źródeł światła:

#### LEDy wykorzystują zjawisko elektroluminescencji do wyemitowania nadwyżki energi w postaci światła spowodowanej przeskakiwaniem elektronów z warstwy N do P w czipie półprzewodnikowym. 

#### Żarówka halogenowa działa na zasadzie podobnej do tradycyjnych modeli żarówek. Prąd wypływa z gniazda i przechodzi przez podstawę żarówki aż do żarnika wolframowego, podgrzewając go. Żarnik zamknięty jest w kwarcowej bańce wypełnionej gazem szlachetnym zmieszanym z niewielką ilością halogenu

#### W świetlówce światło emitowane jest przez luminofor wzbudzony przez promieniowanie ultrafioletowego powstałe na skutek wyładowania jarzeniowego w rurze wypełnionej gazem.

#### W żarówce prąd, przepływając przez cienką spiralę (zwykle wykonaną z wolframu), powoduje jej podgrzanie do temperatury znacznie przekraczającej 2000 K. Wolframowy drucik żarzy się, dając strumień światła.

### Długości fal elektromagnetycznych widocznych dla ludzkiego oka znajdują się w zakresie 380-740 nm. Od długości fali zależy również kolor światła jaki widzimy. Inne parametry widm np. ciągłość wpływają na zastosowanie źródeł światła.

```python
import pandas as pd
```

```python
led_ciepły = pd.read_csv("C:\\Users\\Igor\\Desktop\\Spektro\\Pomiary\\P_led_ciepły.txt", on_bad_lines='skip')
led_zimny = pd.read_csv("C:\\Users\\Igor\\Desktop\\Spektro\\Pomiary\\P_led_zimny.txt", on_bad_lines='skip')
świetlówka_ciepła = pd.read_csv("C:\\Users\\Igor\\Desktop\\Spektro\\Pomiary\\P_świetlówka_ciepła.txt", on_bad_lines='skip')
świetlówka_zimna = pd.read_csv("C:\\Users\\Igor\\Desktop\\Spektro\\Pomiary\\P_świetlówka_zimna.txt", on_bad_lines='skip')
żarówka = pd.read_csv("C:\\Users\\Igor\\Desktop\\Spektro\\Pomiary\\P_żarówka.txt", on_bad_lines='skip')
światłowód = pd.read_csv("C:\\Users\\Igor\\Desktop\\Spektro\\Pomiary\\P_światłowód.txt", on_bad_lines='skip')
halogen = pd.read_csv("C:\\Users\\Igor\\Desktop\\Spektro\\Pomiary\\P_halogen.txt", on_bad_lines='skip')
żółta_fiolka = pd.read_csv("C:\\Users\\Igor\\Desktop\\Spektro\\Pomiary\\P-led zimny żółta fiola-001.txt", on_bad_lines='skip')
różowa_fiolka = pd.read_csv("C:\\Users\\Igor\\Desktop\\Spektro\\Pomiary\\P-led zimny różowa fiolka-001.txt", on_bad_lines='skip')
fioletowa_fiolka = pd.read_csv("C:\\Users\\Igor\\Desktop\\Spektro\\Pomiary\\P_led_zimny_fioletowa_fiolka_001.txt", on_bad_lines='skip')
```

```python
df = pd.concat([led_ciepły,led_zimny,świetlówka_ciepła,świetlówka_zimna,żarówka,światłowód,halogen,żółta_fiolka,różowa_fiolka,fioletowa_fiolka], axis=1)
df.set_axis(['led_ciepły','led_zimny','świetlówka_ciepła','świetlówka_zimna','żarówka','światłowód','halogen','żółta_fiolka','różowa_fiolka', 'fioletowa_fiolka'], axis='columns', inplace=True)
```

```python
df.drop(df.head(9).index, inplace=True)
```

```python
df.head(10)
```

```python
for column in df:
    df[column] = df[column].str.split("\t")
```

```python
df1 = pd.DataFrame(df['led_ciepły'].values.tolist(), columns = ['Wave Length [nm]','Intensity_led_ciepły [a.u.]'])
df2 = pd.DataFrame(df['led_zimny'].values.tolist(), columns = ['Wave Length [nm]','Intensity_led_zimny [a.u.]'])
df3 = pd.DataFrame(df['świetlówka_ciepła'].values.tolist(), columns = ['Wave Length [nm]','Intensity_świetlówka_ciepła [a.u.]'])
df4 = pd.DataFrame(df['świetlówka_zimna'].values.tolist(), columns = ['Wave Length [nm]','Intensity_świetlówka_zimna [a.u.]'])
df5 = pd.DataFrame(df['żarówka'].values.tolist(), columns = ['Wave Length [nm]','Intensity_żarówka [a.u.]'])
df6 = pd.DataFrame(df['światłowód'].values.tolist(), columns = ['Wave Length [nm]','Intensity_światłowód [a.u.]'])
df7 = pd.DataFrame(df['halogen'].values.tolist(), columns = ['Wave Length [nm]','Intensity_halogen [a.u.]'])
df8 = pd.DataFrame(df['żółta_fiolka'].values.tolist(), columns = ['Wave Length [nm]','Intensity_żółta [a.u.]'])
df9 = pd.DataFrame(df['różowa_fiolka'].values.tolist(), columns = ['Wave Length [nm]','Intensity_różowa [a.u.]'])
df10 = pd.DataFrame(df['fioletowa_fiolka'].values.tolist(), columns = ['Wave Length [nm]','Intensity_fioletowa [a.u.]'])
```

```python
dfs = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10]
```

```python
import functools as ft
df_final = ft.reduce(lambda left, right: pd.merge(left, right, on='Wave Length [nm]'), dfs)
```

```python
# wybierz styl wykresu
plt.style.use('seaborn-colorblind')

# utwórz obiekt wykresu
fig, ax = plt.subplots()

ax.set_xlim([350, 920])
ax.set_ylim([0, 250])

Wave_Length = np.asarray(df_final["Wave Length [nm]"], dtype='float')

led_ciepły = np.asarray(df_final["Intensity_led_ciepły [a.u.]"], dtype='float')

led_zimny = np.asarray(df_final['Intensity_led_zimny [a.u.]'], dtype='float')

świetlówka_ciepła = np.asarray(df_final['Intensity_świetlówka_ciepła [a.u.]'], dtype='float')

świetlówka_zimna = np.asarray(df_final['Intensity_świetlówka_zimna [a.u.]'], dtype='float')

żarówka = np.asarray(df_final['Intensity_żarówka [a.u.]'], dtype='float')

światłowód = np.asarray(df_final['Intensity_światłowód [a.u.]'], dtype='float')

halogen = np.asarray(df_final['Intensity_halogen [a.u.]'], dtype='float')

żółta_fiolka = np.asarray(df_final['Intensity_żółta [a.u.]'], dtype='float')

różowa_fiolka = np.asarray(df_final['Intensity_różowa [a.u.]'], dtype='float')

fioletowa_fiolka = np.asarray(df_final['Intensity_fioletowa [a.u.]'], dtype='float')

# dodaj dane
ax.plot(Wave_Length, led_ciepły, label='led_ciepły')
ax.plot(Wave_Length, led_zimny, label='led_zimny')
ax.plot(Wave_Length, świetlówka_ciepła, label='świetlówka_ciepła')
ax.plot(Wave_Length, świetlówka_zimna, label='świetlówka_zimna')
ax.plot(Wave_Length, żarówka, label='żarówka')
ax.plot(Wave_Length, światłowód, label='światłowód')
ax.plot(Wave_Length, halogen, label='halogen ')

# dodaj tytuł i etykiety osi
ax.set_title('Obrazy widm', fontsize = 20)
ax.set_xlabel('Wave_Length [nm]', fontsize = 18)
ax.set_ylabel('Intensity [a.u.]', fontsize = 18)

# dodaj legendę
ax.legend(loc="upper right", bbox_to_anchor=(1.1, 1.1), fontsize = 10, frameon=True)
# wyświetl wykres
plt.show()
```
